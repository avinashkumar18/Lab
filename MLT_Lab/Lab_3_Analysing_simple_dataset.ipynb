```
# importing packages
import numpy as np
from sklearn.datasets import load_breast_cancer
import pandas as pd
```

```
# import cancer dataset
cancer_ds = load_breast_cancer()
```

```
# set the predictor and target variables
X = cancer_ds.data
Y = cancer_ds.target
```

```
print(X)
print(Y)
```

```
data = pd.DataFrame(cancer_ds.data, columns = cancer_ds['feature_names'])
```

```
# setting class column as target variable
data['class'] = cancer_ds.target
```

***Train Test Split***

```
# simple package that splits the train and test data
from sklearn.model_selection import train_test_split
```

```
# dropping the target column
X = data.drop('class', axis=1)
Y = data['class']
```

```
#splitting X and Y
# test size - size of test dataset
# stratify - will giive correct propotion of split among categorical variables(0 and 1 in this case)
# random state - preserves the split shuffling
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.15, stratify=Y, random_state = 2 )
```

```
print(Y.shape, Y_train.shape, Y_test.shape)
```

```
print(X.shape, X_train.shape, X_test.shape)
```

```
print(X.mean(), X_train.mean(), X_test.mean())
```

```
# providing different case below
X_trains, X_tests, Y_trains, Y_tests = train_test_split(X,Y,test_size=0.20, random_state = 5 )
print(Y.shape, Y_trains.shape, Y_tests.shape)
print(X.shape, X_trains.shape, X_tests.shape)
```

```
print(X.mean(), X_trains.mean(), X_tests.mean())
```

```
#Binarization of input
X_binarised_train = X_train.apply(pd.cut, bins=2, labels=[1,0]) 
X_binarised_test = X_test.apply(pd.cut, bins=2, labels=[1,0])
```

```
X_binarised_train = X_binarised_train.values
X_binarised_test = X_binarised_test.values
```

```
print(X_binarised_train) 
print(X_binarised_test)
type(X_binarised_train)
```

```
# predicting the accuracy for every scenario
print("num    acc_rows     accuracy")
for num in range (X_binarised_train.shape[1]+1):
  accurate_rows= 0
  for x,y in zip(X_binarised_train, Y_train):
    y_pred = (np.sum(x)>=num)
    accurate_rows += (y==y_pred)
  print(num,'\t', accurate_rows,'\t',accurate_rows/X_binarised_train.shape[0])
```

```
# above provides row 27 and 28 show most accurate values
# find test accuracy for num = 27 and 28
print("num    acc_rows     accuracy")
for num in range (27,29):
  accurate_rows= 0
  for x,y in zip(X_binarised_train, Y_test):
      y_pred = (np.sum(x)>=num)
      accurate_rows += (y==y_pred)
  print(num, '\t', accurate_rows, '\t', accurate_rows/X_binarised_test.shape[0])
```
