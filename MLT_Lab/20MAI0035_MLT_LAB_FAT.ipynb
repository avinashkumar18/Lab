```python
# Importing necessary packages
import pandas as pd
import numpy as np
```

```python
# Downloaded student performance dataset from https://archive.ics.uci.edu/ml/index.php
# uploaded into my own github repository and used as my dataset
url = "https://raw.githubusercontent.com/avinashkumar18/Lab/main/MLT_Lab/student-mat.csv"
df = pd.read_csv(url)
```

```python
# overview of dataset
df.head()
```

```python
# Dimensions of the dataset
df.shape
```

```python
# Info about the dataset
df.describe
```

Considering Higher as the target value because We are going to predict if the student can able to go for higher studies based on their current performances

```python
# Considering Higher as the target value because
# We are going to predict if the student can able
# to go for higher studies based on their current performances
taget = df['higher']
df = df.drop('higher', axis=1)
```

```python
# One-hot encoding techinique is done to make free of categorical values
df = pd.get_dummies(df)
df.head()
```

```python
taget.head()
```

```python
# Encoding the target value with either 1 or 0
target = taget.map({'yes':1,'no':0})
```

```python
target.value_counts()
```

```python
# checking if a row contain any NA value
df.isna().sum()
```

```python
# The below column is dropped because this column contain only nan values
df = df.drop(['Unnamed: 33'], axis=1)
```

```python
#The last row also contains NA values, so they are also dropped
df = df.drop([395])
target = target.drop([395])
```

```python
df.tail()
```

```python
# Naming the column
target.column=['higher']
```

```python
# Converting datatype of target similar to input datatype
target= target.astype("category")
```

```python
# Assigning the dependent and independent values for model training and pred
x = df
y = target
```

```python
# Importing train test split and splitting the dataset
# 30% goes as test data
from sklearn .model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, random_state = 25)
```

```python
# Since it does not have binary values, we are normalizing the values
X_binarised_train = X_train.apply(pd.cut, bins=2, labels=[1,0]) 
X_binarised_test = X_test.apply(pd.cut, bins=2, labels=[1,0])
```

```python
# Assigning the binary values
X_binarised_train = X_binarised_train.values
X_binarised_test = X_binarised_test.values
```

```python
# IMporting random forest classifier
from sklearn.ensemble import RandomForestClassifier
rc = RandomForestClassifier()
```

I used Random forest because, for this moderate small data, a decision tree can give the best result considering all binary values. The time taken is average compared to KNN and SVM

```python
# fitting the model
rc.fit(X_train, Y_train)
```

```python
predictions = rc.predict(X_test)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
RF_confusion_matrix = confusion_matrix(Y_test, predictions)
RF_confusion_matrix
```

```python
# Accuracy score
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, predictions))
```
