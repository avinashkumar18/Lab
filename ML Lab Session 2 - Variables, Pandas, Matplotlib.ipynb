```python
import numpy as n
```

```python
import pandas as p
data = p.read_csv('MLT\lab\crx.data', header = None)
```

```python
#A1 A2 A3.....A16
varnames = ['A'+ str(s) for s in range(1,17)]
```

```python
#add column names
data.columns = varnames
```

```python
data.head(15)
```

```python
data.tail(15)
```

```python
#? caused my either missing values or encoding issues
# replace ? with not a number using pandas
data = data.replace('?', n.nan)
```

```python
data.info()
```

```python
# recasting A2 and A14
data['A2'] = data['A2'].astype('float')
data['A14'] = data['A14'].astype('float')
```

```python
# encoding A16 values from + and - to 1 and 0
data['A16']  = data['A16'].map({'+':1,'-':0})
```

```python
# Numerical variables
    #Discrete
    #Continous
# Categorical variables
    #Nominals(Names) - R,G,B, OOty 
    #Ordinals(Ordering) - sunday-saturday, 1-10
```

```python
# find categorical variables
cat_columns = [c for c in data.columns if data[c].dtypes=='O']
```

```python
data[cat_columns].head()
```

```python
# find Numerical variables
numerical_columns = [nu for nu in data.columns if data[nu].dtypes!='O']
```

```python
data[numerical_columns].head()
```
