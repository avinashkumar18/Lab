```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
for pic in range(1,4):
    image=mpimg.imread(r'DC dataset/superman/'+str(pic)+'.jpg')
    plt.figure()
    plt.imshow(image)
```

```python
image_list=[]
target=[]
image_dataset=[]
```

```python
for pic in range(1,7):  
    path=r'DC dataset/superman/'+str(pic)+'.jpg'
    load=Image.open(path)
    resize_img = load.resize((np.array((250,250))).astype(int))
    input_img = np.asarray(resize_img)
    input_img=input_img.flatten()
    image_list.append(input_img)
    target.append(0)
superman_symbol=np.array(image_list)
for s in range(6):
    image_dataset.append(list(superman_symbol[s]))
```

```python
image_list=[]
for pic in range(1,3):
    path=r'DC dataset/batman/'+str(pic)+'.jpg'
    load=Image.open(path)
    resize_img = load.resize((np.array((250,250))).astype(int))
    input_img = np.asarray(resize_img)
    input_img=input_img.flatten()
    image_list.append(input_img)
    target.append(1)
batman_symbol=np.array(image_list)
for b in range(2):
    image_dataset.append(list(batman_symbol[b]))
```

```python
X_train=np.array(image_dataset)
Y_train=np.array(target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train,Y_train)
```

```python
X_test=[]
Y_test=[]
for pic in range(2):
    path=r'DC dataset/superman/'+str(6+pic)+'.jpg'
    load=Image.open(path)
    resize_img = load.resize((np.array((250,250))).astype(int))
    input_img = np.asarray(resize_img)
    input_img=input_img.flatten()
    X_test.append(list(input_img))
    Y_test.append(1)
    
    path=r'DC dataset/batman/'+str(1+pic)+'.jpg'
    load=Image.open(path)
    resize_img = load.resize((np.array((250,250))).astype(int))
    input_img = np.asarray(resize_img)
    input_img=input_img.flatten()
    X_test.append(list(input_img))
    Y_test.append(0)
X_test=np.array(X_test)
Y_test=np.array(Y_test)

Y_Predction = classifier.predict(X_test)
```

```python
print(confusion_matrix(Y_test,Y_Predction))
print(classification_report(Y_test,Y_Predction))
```
