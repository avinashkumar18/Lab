```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10, 10)
```

**Sigmoid function**

fn(x) = 1/(1+exp(-x))

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
plt.plot(x, sigmoid(x))
plt.title('Activation Function :Sigmoid')
plt.show()
```

**Tanh function**

fn(x) = (1-exp(-2x))/(1+exp(-2x))

```python
def tanh(x):
    return np.tanh(x)
```

```python
plt.plot(x, tanh(x))
plt.title('Activation Function :Tanh')
plt.show()
```

**Relu function**

f(x) = max(x,0)


```python
def relu(x):
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)
    return x1
```

```python
plt.plot(x, relu(x))
plt.title('Activation Function :RELU')
plt.show()
```

**Leaky-Relu function**

fn(x) = max(0.01*x,x)

```python
def lrelu(x):
  if x > 0:
    return x
  else:
    return x*0.01
```

```python
plt.plot(x, list(map(lambda x: lrelu(x),x)))
plt.title('Activation Function :Leaky-RELU')
plt.show()
```

**Exponential learning Unit ELu**

f(x) = x if x>0, else a((exp(x))1)

```python
def elu(x):
    if x > 0 :
        return x
    else : 
        return (np.exp(x)-1)
```

```python
plt.plot(x, list(map(lambda x: elu(x),x)))
plt.title('Activation Function :ELU')
plt.show()
```
